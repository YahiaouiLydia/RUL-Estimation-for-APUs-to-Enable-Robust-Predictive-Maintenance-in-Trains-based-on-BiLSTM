```python
import pandas as pd
MetroPT1= pd.read_csv(r'/content/drive/MyDrive/Colab Notebooks/Conference 2024/Datasets/MetroPT1.csv')
MetroPT3 = pd.read_csv(r'/content/drive/MyDrive/Colab Notebooks/Conference 2024/Datasets/MetroPT3.csv')
```

```python
columns_to_remove1 = ['Flowmeter', 'gpsQuality', 'gpsLat', 'gpsSpeed', 'gpsLong']
MetroPT1=MetroPT1.drop(columns=columns_to_remove1)
columns_to_remove3 = ['Unnamed: 0']
MetroPT3 = MetroPT3.drop(columns=columns_to_remove3)
```

```python
# Concaténer verticalement
MetroPT = pd.concat([MetroPT1, MetroPT3], ignore_index=True)
```

```python
import datetime

#Check the current type of timestamp
print(f"Current type of timestamp is {type(MetroPT.timestamp[0])}")

#Convert timestamp to pandas.DateTime
MetroPT['timestamp'] = MetroPT['timestamp'].apply(pd.to_datetime, format = "%Y-%m-%d %H:%M:%S")

#Re-check the type
print(f"Current type of timestamp is {type(MetroPT.timestamp[0])}")
```

```python

# Définir les plages de temps et les valeurs associées
time_ranges = [
    ('2022-02-28 21:53', '2022-03-01 02:00', 'Air Leak'),
    ('2022-03-23 14:54', '2022-03-23 15:24', 'Air Leak'),
    ('2022-05-30 12:00', '2022-06-02 06:18', 'Oil Leak'),
    ('2020-04-18 00:00', '2020-04-18 23:59', 'Air Leak'),
    ('2020-05-29 23:30', '2020-05-30 06:00', 'Air Leak'),
    ('2020-06-05 10:00', '2020-06-07 14:30', 'Air Leak'),
    ('2020-07-15 14:30', '2020-07-15 19:00', 'Air Leak')
]

# Ajouter la colonne 'state' avec une valeur par défaut
MetroPT['state'] = 'Normal'

# Appliquer les plages de temps et les valeurs associées
for start, end, state in time_ranges:
    start_time = pd.to_datetime(start)
    end_time = pd.to_datetime(end)
    mask = (MetroPT['timestamp'] >= start_time) & (MetroPT['timestamp'] <= end_time)
    MetroPT.loc[mask, 'state'] = state

print("DataFrame avec la colonne 'state' ajoutée :")
print(MetroPT)
```

```python


# Définir les plages de temps et les valeurs associées
time_ranges = [
    ('2022-02-28 21:53', '2022-03-01 02:00', 'Clients'),
    ('2022-03-23 14:54', '2022-03-23 15:24', 'Air Dryer'),
    ('2022-05-30 12:00', '2022-06-02 06:18', 'Compressor'),
    ('2020-04-18 00:00', '2020-04-18 23:59', 'Compressor'),
    ('2020-05-29 23:30', '2020-05-30 06:00', 'Compressor'),
    ('2020-06-05 10:00', '2020-06-07 14:30', 'Compressor'),
    ('2020-07-15 14:30', '2020-07-15 19:00', 'Compressor')
]

# Ajouter la colonne 'failure_component' avec une valeur par défaut
MetroPT['failure_component'] = 'Operational'

# Appliquer les plages de temps et les valeurs associées
for start, end, component in time_ranges:
    start_time = pd.to_datetime(start)
    end_time = pd.to_datetime(end)
    mask = (MetroPT['timestamp'] >= start_time) & (MetroPT['timestamp'] <= end_time)
    MetroPT.loc[mask, 'failure_component'] = component

print("DataFrame avec la colonne 'failure_component' ajoutée :")
print(MetroPT)
```

```python
# Définir les plages de temps pour les défaillances
failure_times = [
    ('2022-02-28 21:53', '2022-03-01 02:00'),
    ('2022-03-23 14:54', '2022-03-23 15:24'),
    ('2022-05-30 12:00', '2022-06-02 06:18'),
    ('2020-04-18 00:00', '2020-04-18 23:59'),
    ('2020-05-29 23:30', '2020-05-30 06:00'),
    ('2020-06-05 10:00', '2020-06-07 14:30'),
    ('2020-07-15 14:30', '2020-07-15 19:00')
]

# Ajouter une colonne 'RUL' initialement à 0
MetroPT['RUL'] = 0

# Calculer RUL pour chaque plage de défaillance
for start, end in failure_times:
    # Convertir les chaînes en DateTime
    start_time = pd.to_datetime(start)
    end_time = pd.to_datetime(end)

    # Créer un masque pour les enregistrements dans cette plage de défaillance
    mask = (MetroPT['timestamp'] >= start_time) & (MetroPT['timestamp'] <= end_time)

    # Mettre à jour RUL pour les enregistrements dans cette plage de défaillance à 0
    MetroPT.loc[mask, 'RUL'] = 0

# Pour les enregistrements en dehors de toutes les plages de défaillance, calculer RUL en fonction de la dernière date de défaillance
latest_failure_time = pd.to_datetime(max(end for _, end in failure_times))
outside_failure_mask = MetroPT['RUL'] == 0
MetroPT.loc[outside_failure_mask, 'RUL'] = (latest_failure_time - MetroPT.loc[outside_failure_mask, 'timestamp']).dt.total_seconds() / (60 * 60)

print("DataFrame avec la colonne 'RUL' ajoutée :")
print(MetroPT)
```
